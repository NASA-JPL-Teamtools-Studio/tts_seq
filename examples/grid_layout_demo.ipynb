# IPythonGrid Layout Demo

This notebook demonstrates how to use the `IPythonGrid` class with percentage and pixel-based sizing for cells.

```python
import sys
import os
# Add the src directory to the path so we can import the module
sys.path.append(os.path.abspath('../src'))

from tts_seq.sim_visualization.layout import IPythonCell, IPythonRow, IPythonGrid
from tts_seq.sim_visualization.grid_example import create_sample_plot, create_sample_widget
```

## Basic Grid with Percentage-based Sizing

Create a simple 2x2 grid with percentage-based sizing for both rows and columns.

```python
# Create some sample content
plot1 = create_sample_plot("Plot 1")
plot2 = create_sample_plot("Plot 2")
widget1 = create_sample_widget("This is a widget with some text content.")
widget2 = create_sample_widget("Another widget with different content.")

# Create cells with content
cell1 = IPythonCell(plot1)
cell2 = IPythonCell(widget1)
cell3 = IPythonCell(plot2)
cell4 = IPythonCell(widget2)

# Create rows with cells
row1 = IPythonRow([cell1, cell2])
row2 = IPythonRow([cell3, cell4])

# Create grid with rows
percentage_grid = IPythonGrid([row1, row2])

# Configure layout with percentages
percentage_grid.configure_layout(
    row_heights=['50%', '50%'],   # Equal height rows
    column_widths=['70%', '30%']  # 70/30 split for columns
)

# Display the grid
percentage_grid
```

## Grid with Pixel-based Sizing

Create a 2x2 grid with pixel-based sizing for both rows and columns.

```python
# Create cells with content
cell1 = IPythonCell(create_sample_plot("Fixed Size Plot 1"))
cell2 = IPythonCell(create_sample_widget("Fixed width widget"))
cell3 = IPythonCell(create_sample_plot("Fixed Size Plot 2"))
cell4 = IPythonCell(create_sample_widget("Another fixed width widget"))

# Create rows with cells
row1 = IPythonRow([cell1, cell2])
row2 = IPythonRow([cell3, cell4])

# Create grid with rows
pixel_grid = IPythonGrid([row1, row2])

# Configure layout with pixel sizes
pixel_grid.configure_layout(
    row_heights=['300px', '300px'],    # Fixed height rows
    column_widths=['600px', '300px']   # Fixed width columns
)

# Display the grid
pixel_grid
```

## Mixed Sizing (Percentages and Pixels)

Create a grid with a mix of percentage and pixel-based sizing.

```python
# Create cells with content
cell1 = IPythonCell(create_sample_plot("Main Plot"))
cell2 = IPythonCell(create_sample_widget("Control Panel"))
cell3 = IPythonCell(create_sample_widget("Data Table"))
cell4 = IPythonCell(create_sample_widget("Status Information"))

# Create rows with cells
row1 = IPythonRow([cell1, cell2])
row2 = IPythonRow([cell3, cell4])

# Create grid with rows
mixed_grid = IPythonGrid([row1, row2])

# Configure with mixed sizing
mixed_grid.configure_layout(
    row_heights=['400px', '200px'],     # Fixed heights for rows
    column_widths=['70%', '30%']        # Percentage widths for columns
)

# Display the grid
mixed_grid
```

## Complex Layout Example

Create a more complex layout with different numbers of cells per row.

```python
# Create a grid with varying number of cells per row
header_cell = IPythonCell(create_sample_widget("<h2>Dashboard Header</h2>"))
header_row = IPythonRow([header_cell])

plot1 = IPythonCell(create_sample_plot("Main Plot"))
plot2 = IPythonCell(create_sample_plot("Secondary Plot"))
plots_row = IPythonRow([plot1, plot2])

widget1 = IPythonCell(create_sample_widget("Data Table 1"))
widget2 = IPythonCell(create_sample_widget("Data Table 2"))
widget3 = IPythonCell(create_sample_widget("Data Table 3"))
widgets_row = IPythonRow([widget1, widget2, widget3])

footer_cell = IPythonCell(create_sample_widget("<p>Dashboard Footer</p>"))
footer_row = IPythonRow([footer_cell])

# Create the grid
complex_grid = IPythonGrid([header_row, plots_row, widgets_row, footer_row])

# Configure the layout
complex_grid.configure_layout(
    row_heights=['80px', '400px', '200px', '50px'],
    column_widths=['50%', '50%']  # This will apply to the rows with 2 columns
                                   # Rows with 1 column will span 100%
                                   # Rows with 3 columns will be distributed proportionally
)

# Display the grid
complex_grid
```
